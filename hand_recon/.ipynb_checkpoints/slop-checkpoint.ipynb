```python
import numpy as np
import math
def getValue(x, y, img):
    value = 0
    slope = find_m(x, y, img)#第一個元素是 m, 第二個是值
    if slope[0] == 0:#直接計算
        value = slope[1]
    else:
        value = sin(math.atan(slope[0])) * slope[1]
    return value

def find_m(x, y, img):
    ori = img[y][x]
    solution = [0,0]
    width = 0
    m = 0
    for i in range(1, img.shape[1]): #right
        if ori != img[y][x+i]:
            if img[y-(img.shape[0]/20)][x+i] == img[y][x+i]:
                x_shift = __left__(x+i, y-(img.shape[0]/20), img)
                if x_shift != 0:
                    m = -1 * (y-(img.shape[0]/20)/x_shift)
            else:
                m = Lfind_m(x, y, img)
            width+=i
            break
        ori = img[y][x+i]
    ori = img[y][x]
    for i in range(1, img.shape[1]): #left
        if ori != img[y][x-i]:
            width+=i
            break
        ori = img[y][x-i]
    solution[0] = m
    solution[1] = width
    return solution
def Lfind_m(x, y, img):
    ori = img[y][x]
    m = 0
    for i in range(1, img.shape[1]): #left
        if ori != img[y][x-i]:
            if img[y-(img.shape[0]/20)][x-i] == img[y][x-i]:
                x_shift = __right__(x-i, y-(img.shape[0]/20), img)
                if x_shift != 0:
                    m = (y-(img.shape[0]/20)/x_shift)
            break
    ori = img[y][x-i]
    return m
def __left__(x, y, img):
    for i in range(1, img.shape[1]): #left
        if ori != img[y][x-i]:
            return i
    return 0
def __right__(x, y, img):
    for i in range(1, img.shape[1]): #right
        if ori != img[y][x+i]:
            return i
    return 0
```
